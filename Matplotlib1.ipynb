```python
#plotting two time series together
fig, ax = plt.subplots()
ax.plot(cl.index, cl["co2"])
ax.plot(cl.index, cl["relative_temp"])
ax.set_xlabel("time")
ax.set_ylabel("co2/relative temp")
ax.set_title("co2/relative temp over time")
```

```python
fig, ax = plt.subplots()
ax.plot(cl.index, cl['co2'],color="blue")
ax.set_xlabel("time")
ax.set_ylabel("co2",color="blue")
ax.tick_params('y',colors="blue")

ax2 = ax.twinx()

ax2.plot(cl.index, cl["relative_temp"],color="red")
ax2.set_ylabel("relative_temp",color="red")
ax2.tick_params('y',colors="red")
```

```python
def plot_timeseries(axes, x, y, color, xlabel,ylabel):
    axes.plot(x,y,color=color)
    axes.set_xlabel(xlabel)
    axes.set_ylabel(ylabel)
    axes.tick_params('y',colors=color)
```

```python
fig, ax = plt.subplots()
plot_timeseries(ax, cl.index, cl["co2"], color="blue", xlabel="time", ylabel="co2")
ax2 = ax.twinx()
plot_timeseries(ax2, cl.index, cl["relative_temp"], color="red", xlabel="time", ylabel="relative temp")

ax2.annotate(
    text=">1",
    xy=(pd.Timestamp('2015-10-06'),1),
    xytext=(pd.Timestamp('2008-10-06'),-0.2),
    arrowprops={'arrowstyle':'->', 'color':'gray'}
)
plt.show()
```

```python
medals = pd.read_csv("Datasets/medals.csv",index_col=0)
medals.head()
```

```python
fig, ax = plt.subplots()
ax.bar(medals.index, medals["Gold"])
ax.set_xticklabels(medals.index, rotation=90)
ax.set_ylabel("medals")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(medals.index, medals["Gold"])
ax.bar(medals.index, medals["Silver"], bottom=medals["Gold"])
ax.bar(medals.index, medals["Bronze"], bottom=medals["Gold"]+medals["Silver"])
ax.set_ylabel("Medals")
ax.set_xticklabels(medals.index, rotation=90)
plt.legend(["Gold", "Silver","Bronze"])
plt.show()
```

```python
#gymnastics
gym = pd.read_csv("Datasets/medalsweight.csv",index_col=0)
gym.head(2)

gym_rowing = gym[gym["Sport"] == "Rowing"]
gym_gymnastics = gym[gym["Sport"] == "Gymnastics"]
```

```python
fig, ax = plt.subplots()
ax.bar('Rowing', gym_rowing["Height"].mean())
ax.bar('Gymnastics', gym_gymnastics["Height"].mean())
plt.show()
```

```python
fig, ax = plt.subplots()
ax.hist(gym_rowing["Height"],label="Rowing",histtype="step")
ax.hist(gym_gymnastics["Height"],label="Gymnastics",histtype="step")
ax.set_ylabel("Height")
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots()

fig.set_size_inches([8,6])

ax.bar('Rowing', gym_rowing["Height"].mean(),
      yerr=gym_rowing["Height"].std())

ax.bar('Gymnastics', gym_gymnastics["Height"].mean(),
      yerr=gym_rowing["Height"].std())

plt.show()
```

```python
fig, ax = plt.subplots()

ax.boxplot([gym_rowing["Height"], gym_gymnastics["Height"]])
ax.set_xticklabels(['Rowing','Gymnastics'])
```

```python
#scatterplot

cl.index = pd.to_datetime(cl.index)

fig,ax = plt.subplots()

ax.scatter(
            cl['co2'], 
           cl['relative_temp'], 
           c=cl.index
)
ax.set_xlabel("co2")
ax.set_ylabel("relative temp")
ax.set_title("co2 vs relative temperature")
plt.show()
```

```python
eighties = cl['1980-01-01':'1989-12-31']
nintees = cl['1990-01-01':'1999-12-31']

fig, ax = plt.subplots()
ax.scatter(eighties['co2'],eighties['relative_temp'],color='red',label="eightees")
ax.scatter(nintees['co2'],nintees['relative_temp'],color='blue',label="ninetees")

plt.legend()
plt.show()

fig.savefig("co2_vs_relativetemp.png",dpi=200)
```
