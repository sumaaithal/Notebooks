```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
years = [1800, 1850, 1900] + list(range(1950, 2100))
population =  [1.0, 1.262, 1.650] + [round(num,2) for num in np.linspace(2.5,12, len(years)-3)]
```

```python
plt.scatter(x=years, y=population)
plt.xlabel("years")
plt.ylabel("population")
plt.title("Population projections")
plt.yticks(ticks=[0,2,4,6,8,10,12], labels=['0', '2B', '4B', '6B', '8B','10B', '12B'])
plt.show()
```
