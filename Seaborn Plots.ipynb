```python
ctr = pd.read_csv("Datasets/Countries.csv")
ctr.head()
```

```python
ctr['Birthrate'] = ctr['Birthrate'].str.replace(",",".")
ctr['Birthrate'] = ctr['Birthrate'].astype(float)
```

```python
sns.set_context("paper")

g = sns.catplot(data=ctr, x="Region",y="Birthrate",kind="box")

g.fig.suptitle("Region Vs BirthRate",y=1.03)

g.set(xlabel="Region",ylabel="Birth Rate")

plt.xticks(rotation=90)

plt.show()
```
