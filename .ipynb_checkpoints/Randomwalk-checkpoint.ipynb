```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(123)
final_walk = []

for x in range(10000):
    tails = [0]
    
    for x in range(10):
        coins = np.random.randint(0,2)
        tails.append(tails[x]+coins)

    final_walk.append(tails[-1])
```

```python
plt.hist(final_walk, bins=10)
plt.show()
```
